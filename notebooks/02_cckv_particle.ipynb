```python
import warnings
warnings.filterwarnings('ignore')
from cherenkov_toy import Particle, Medium, CherenkovPerParticle, CherenkovMontecarlo, c
import numpy as np
```

### Creating a particle configuration

```python
electron = Particle(0.9 * c, 1, 1)
strange_medium = Medium(lambda omega: np.sqrt(omega / 1e15))
cherenkov = CherenkovPerParticle(electron, strange_medium)
omegas = np.linspace(1e6, 1e16, 1000)
cherenkov.cherenkov_angle_range_degrees_plot(omegas)
```
