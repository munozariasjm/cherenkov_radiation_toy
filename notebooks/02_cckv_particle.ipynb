```python
import warnings
warnings.filterwarnings('ignore')
from cherenkov_toy import Particle, Medium, CherenkovPerParticle, CherenkovMontecarlo, c
import numpy as np
```

### Creating a particle configuration

```python
electron = Particle(0.9 * c, 1, 1)
strange_medium = Medium(lambda omega: np.sqrt(omega / 1e15))
cherenkov = CherenkovPerParticle(electron, strange_medium)
omegas = np.linspace(1e6, 1e16, 1000)
cherenkov.cherenkov_angle_range_degrees_plot(omegas)
```

We note that this phenomenon is a direct consequence of the strange definition of the speed of light in a medium. The speed of light in a medium is defined as the speed of light in vacuum divided by the refractive index of the medium. This means that the speed of light in a medium is always less than the speed of light in vacuum. This is why the speed of light in a medium is also called the phase velocity of light in that medium. However, for a constant speed as a function of the frecuency, we don't observe this in our code:

```python
water = Medium(refractive_index=1.33)
cherenkov = CherenkovPerParticle(electron, water)
cherenkov.cherenkov_angle_range_degrees_plot(omegas)
```
